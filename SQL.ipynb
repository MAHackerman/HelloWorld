# SQL and BigQuery (Kaggle)
SQL (Structured Query Language)is a programming language which is used for relational databases. 
BigQuery is a web based service that will let you apply SQL to databases.
First you create a client object. In kaggle, the public dataset "BigQuery" is used. However, it is not open source, so I'll just copy and paste the BigQuery methods used in the Kaggle excercise

The structure is the following: 
Project > Datasets > Tables


The structure of a table is called <i>schema</i>.
Each SchemaField tells us information about a specific column (also called field) in a specific order:
1. column (field) name
2. field type (or datatype in the column, e.g. string)
3. The mode of the column ('NULLABLE' means that a column allows NULL values, and is the default)
4. A description of the data in that column

You always need to first construct a reference (first to a dataset and then to a table). 
Use the get_datset (or get_table) method to fetch the dataset (or table)
